# 목표

이 노트북은 PyTorch와 Ray Tune을 활용한 CIFAR10 이미지 분류기 학습 및 하이퍼파라미터 튜닝을 구현하기 위한 실습 파일입니다. 아래 각 셀의 지시사항에 따라 빈 셀(Code)을 추가하여 코드를 작성하세요.

**[Step 1: 라이브러리 설치 및 환경 설정]**

- 환경 변수 `RAY_TRAIN_V2_ENABLED`를 `"1"`로 설정하세요.
- 필요한 패키지를 설치하는 명령어를 작성하세요:
  설치할 패키지: `ray[tune]`, `ray[train]`, `torch`, `torchvision`, `matplotlib`, `ipywidgets`

**[Step 2: 모듈 임포트]**

- 기본 파이썬 라이브러리: `os`, `tempfile`, `tqdm.auto.tqdm`
- 데이터 처리: `numpy`, `filelock.FileLock`
- 머신러닝 라이브러리: `torch` (nn, functional, optim), `torchvision` (transforms, datasets)
- 데이터 유틸리티: `torch.utils.data.random_split`
- Ray Tune 관련: `ray.tune`, `ray.tune.schedulers.ASHAScheduler`

**[Step 3: 데이터 로더 구현]**

다음 세 가지 함수를 작성하여 데이터를 준비하세요.
- `load_data(data_dir)`: CIFAR10 데이터를 로드하고 변환(Transform)합니다. (`FileLock` 사용 권장)
- `create_dataloaders(trainset, batch_size, num_workers)`: train 데이터셋을 train/val 서브셋으로 분할하고 각각의 데이터로더를 반환합니다.
- `load_test_data()`: 테스트용 가짜 데이터(FakeData)를 로드합니다.

**[Step 4: 신경망 정의]**

- `nn.Module`을 상속받는 `Net` 클래스를 구현하세요.
- 초기화(`__init__`) 시 신경망의 계층 크기(예: `l1`, `l2`)를 외부에서 설정할 수 있도록 인자로 받으세요.

**[Step 5: 학습 함수 구현]**

- `config`를 인자로 받는 `train_cifar(config)` 함수를 작성하세요.
- 모델 초기화 및 장치(device) 할당, 손실 함수와 옵티마이저를 설정합니다.
- 에폭마다 모델을 학습하고 검증(Validation)을 수행합니다.
- `tune.report`를 통해 메트릭(loss, accuracy)과 체크포인트를 Ray Tune에 보고하세요.

**[Step 6: 최고 성능 모델 테스트 함수 구현]**

- 튜닝 결과(`best_result`)를 인자로 받아 가장 성능이 좋은 모델을 테스트하는 `test_best_model(best_result)` 함수를 작성하세요.
- 최고 성능의 체크포인트에서 가중치를 불러오고, 테스트 데이터셋을 통해 최종 정확도를 측정하여 출력합니다.

**[Step 7: 하이퍼파라미터 탐색 공간 설정 및 실행]**

- 학습률(`lr`), 배치 사이즈(`batch_size`), 레이어 크기(`l1`, `l2`) 등의 하이퍼파라미터 탐색 공간(`config` 딕셔너리)을 정의하세요.
- `ASHAScheduler`와 `tune.Tuner`를 설정하고 하이퍼파라미터 튜닝을 실행하는 `main` 함수를 구현하세요.
- `ray.init()`를 호출하여 임시 Ray 클러스터를 생성하여 활용합니다.
- `main` 함수를 직접 호출하여 전체 과정을 실행하세요. (예: `main(config, gpus_per_trial=1)`)

**[Step 8: 결과 시각화]**

- `matplotlib.pyplot` 라이브러리를 임포트하세요.
- 최고 성능으로 학습된 모델(best model)을 불러옵니다.
- 테스트 데이터셋(Test data) 중 일부 이미지에 대해 예측(prediction)을 수행하세요.
- 원본 이미지와 함께 예측된 라벨(Predicted Label), 실제 라벨(Actual Label)이 나타나도록 시각화(plot) 코드를 작성하세요.
